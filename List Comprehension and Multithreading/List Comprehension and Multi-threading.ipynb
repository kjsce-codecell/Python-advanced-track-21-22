### Dictionary Comprehension

```python
keys = ['x', 'y', 'z']
values = [1, 2, 3]

dictionary = {k:v for k, v in zip(keys, values)}
print(list(zip(keys, values)))
print(dictionary)
```

### Multi-Threading

```python
from time import *
import threading
```

```python
def hello(range1):
    for _ in range(range1):
        print("hello")
        sleep(1)

def hi(range2):
    for _ in range(range2):
        print("hi")
        sleep(1)


t1 = threading.Thread(target = hello, args = [10])
t2 = threading.Thread(target = hi, args = [10])

t1.start()
sleep(0.4)
t2.start()

t1.join()
t2.join()

print("====================================")
```

```python
def even(number1):
  if(number1 % 2 == 0):
    print("Even")
    sleep(0.2)

def odd(number2):
  if(number2 % 2 != 0):
    print("Odd")
    sleep(0.2)

t1 = threading.Thread(target = even, args = [int(input("Enter a number: "))])
t2 = threading.Thread(target = odd, args = [int(input("Enter a number: "))])

t1.start()
sleep(0.4)
t2.start()

t1.join()
t2.join()

print("====================================================================")
```

```python
def list_comp(power, upper_range):
  l = [i ** power for i in range(0, upper_range)]
  print(l)
  sleep(0.3)

def print_nos(no):
  for i in range(no):
    print(i)

t1 = threading.Thread(target = list_comp, args = [int(input("Enter the power: ")), int(input("Enter the upper range: "))])
t2 = threading.Thread(target = print_nos, args = [int(input("Enter range of numbers: "))])

t1.start()
sleep(0.2)
t2.start()

t1.join()
t2.join()

print("============================================================")
```
